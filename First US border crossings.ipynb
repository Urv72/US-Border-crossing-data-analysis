```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Border_Crossing_Entry_Data1.csv',sep=',')
```

```python
df
```

```python
df.groupby('Port Name')['Value'].sum()
```

```python
df.groupby('Port Name')['Value'].sum().sort_values(ascending=False)
```

```python
# ploting series in the form of bar plot
ax = df.groupby('Port Name')['Value'].sum().sort_values(ascending=False).head(10).plot(kind='bar',figsize=(17,6))

# setting title for bar plot
ax.set_title('Port with Highest Border crossing entries')

# printing values on top of every bar
for i in ax.patches:
    ax.text(i.get_x()-.2,i.get_height()+10000000,str(i.get_height()),fontsize=15,color='black') 

# for vertical bar plot :       ax.text(i.get_x(),i.get_height(),str(i.get_height()))
# for horizontal bar plot :       ax.text(i.get_y(),i.get_width(),str(i.get_width()))
# where ,
# i.get_x() and i.get_y() used for left-right positioning
# i.get_height() and i.get_width() user for up-down positioning
```

```python
ax = df.groupby('Port Name')['Value'].sum().sort_values(ascending=False).tail(10).plot(kind='bar',figsize=(17,6))

ax.set_title('Port with Lowest border crossing entries')

for i in ax.patches:
    ax.text(i.get_x(),i.get_height()+2000,str(i.get_height()),fontsize=15,color='black')
```

```python
df.groupby('Year')['Value'].sum().iloc[0:10]
```

```python
ax = df.groupby('Year')['Value'].sum().iloc[0:13].plot(kind='bar',figsize=(20,8))

ax.set_title('Border crossing entries Yearwise',fontsize=25)
ax.set_xlabel('Year',fontsize=15)


for i in ax.patches:
    ax.text(i.get_x()-.2,i.get_height()+5000000,str(i.get_height()),fontsize=15,color='black')
```
