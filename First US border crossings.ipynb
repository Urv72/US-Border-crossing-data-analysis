```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Border_Crossing_Entry_Data1.csv',sep=',')
```

```python
df
```

```python
df.groupby('Port Name')['Value'].sum().sort_values(ascending=False)
```

```python
df.groupby('Port Name')['Value'].sum().sort_values(ascending=False).head(10).plot(kind='bar',figsize=(12,6)).set_title('Port with Highest Border crossing entries')
```

```python
df.groupby('Port Name')['Value'].sum().sort_values(ascending=False)
```

```python
df.groupby('Port Name')['Value'].sum().sort_values(ascending=False).tail(10).plot(kind='bar',figsize=(12,6)).set_title('Port with Lowest border crossing entries')
```

```python
df.groupby('State')['Value'].sum().sort_values(ascending=False)
```

```python
df.groupby('State')['Value'].sum().sort_values(ascending=False).head(10).plot(kind='bar',figsize=(12,6)).set_title('State in demand among the border crossers')
```

```python
df.groupby('Border')['Value'].sum().sort_values(ascending=False)
```

```python
df.groupby('Border')['Value'].sum().sort_values(ascending=False).plot(kind='bar').set_title('Border vs Value') # .text(0.3,1,str(v), color='blue', fontweight='bold')
```

```python
list1 = df['Date'].to_list()
```

```python
for i in range(0,len(list1)):
    list1[i] = list1[i][6:10]
```

```python
df.groupby('Year')['Value'].sum().sort_index()
```
